```python
#importing all the important libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#loading the dataset into a dataframe
df=pd.read_csv("house price.csv")
```

```python
#understanding the data

print("CHECKING SHAPE OF THE DATASET")
print("No. of rows in the dataset:",df.shape[0])
print("No. of columns in the dataset:",df.shape[1])

print("\nNAME OF ALL THE COLUMNS")
print(df.columns)

print("\nUNDERSTANDING DATATYPE OF EACH COLUMN AND NON-NULL COUNTS OF EACH COLUMN")
df.info()

print("\nUNDERSTANDING SUMMARY STATISTICS OF EACH COLUMN")
df.describe()
```

```python
# removing unwanted columns from our dataset
df.drop(['view', 'sqft_lot'], axis=1, inplace=True)
df.drop(['statezip'], axis=1, inplace=True)
```

```python
# Plotting the histogram

# Set the figure size
plt.figure(figsize=(6,4),facecolor="#faedcd")  
plt.hist(df['sqft_living'],bins=8, color='#d4a373', edgecolor='#bc6c25',histtype="barstacked",align="left")
plt.gca().set_facecolor("#fefae0")
#setting x-axis
plt.xlim(0,8000)
plt.xticks(range(0,8000,1000))

# Add titles and labels
plt.title('Distribution of House Area', fontsize=16)
plt.xlabel('Area in sqft', fontsize=12)
plt.ylabel('Number of Houses', fontsize=12)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(6,4),facecolor="#ffdab9")
m={0:"o",1:">"}
custom = sns.color_palette(["#f08080", "#9e2a2b"])
graph=sns.scatterplot(x="sqft_living",y="price",data=df,hue="waterfront",palette=custom,size="waterfront",sizes=(100,50),
                      alpha=0.7,style="waterfront",markers=m)
plt.gca().set_facecolor("#fefae0")
graph.set_title("Relationship Between Price and Living Area (sqft)",fontsize=15)
graph.set_xlabel("Living area(sqft)",fontsize=12)
graph.set_ylabel("Price",fontsize=12)
plt.plot()
```

```python
plt.figure(facecolor="#caf0f8")
condition1=df.price[df.condition==1]
condition2=df.price[df.condition==2]
condition3=df.price[df.condition==3]
condition4=df.price[df.condition==4]
condition5=df.price[df.condition==5]
a=len(condition1)
b=len(condition2)
c=len(condition3)
d=len(condition4)
e=len(condition5)
z=[a,b,c,d,e]
label=[1,2,3,4,5]
c=["#1d3557","#457b9d","#48cae4","#219ebc","#2a9d8f"]
plt.pie(z,labels=label,explode=[0.1,0.3,0.05,0.05,0.05],labeldistance=1.05,autopct="%d%%",shadow=False,colors=c)
plt.title("Distribution of Houses Based on Condition",fontsize=15)
plt.plot()
```

```python
plt.figure(facecolor="#d4a276",figsize=(6,5))
a1=np.average(condition1)
a2=np.average(condition2)
a3=np.average(condition3)
a4=np.average(condition4)
a5=np.average(condition5)
data1=[a1,a2,a3,a4,a5]
plt.plot([1,2,3,4,5],data1,marker='o',markerfacecolor="#6c584c",markeredgecolor="#772e25",linestyle="-.",color="#d4a373")
plt.gca().set_facecolor("#faedcd")
plt.xticks(range(1,6,1))
plt.title("Relation b/w Price & Condition",fontsize=15)
plt.xlabel("Condition",fontsize=12)
plt.ylabel("Average Price",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(11,4),facecolor="#e0e1dd")
graph1=sns.countplot(x="city",data=df,width=0.5,palette="deep",hue="city",legend=False)
plt.gca().set_facecolor("#fdfcdc")
graph1.set_title("No. of houses per city",fontsize=15)
graph1.set_xlabel("City",fontsize=12)
graph1.set_ylabel("No. of houses",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(11,5),facecolor="#adc178")
plt.bar(df.city,df.price,width=0.5,color="#588157")
plt.gca().set_facecolor("#dde5b6")
plt.title("Average price of house per city",fontsize=15)
plt.xlabel("City",fontsize=12)
plt.ylabel("Average price",fontsize=12)
plt.show()
df.head()
```

```python
plt.figure(figsize=(6,5),facecolor="#ffef9f")
custom1 = sns.color_palette(["#22577a","#8ecae6","#f5cb5c","#f4acb7","#b388eb"])
sns.stripplot(x="bedrooms",y="price",data=df,hue="condition",palette=custom1,size=8,alpha=0.9)
plt.gca().set_facecolor("#fcfffc")
plt.title("Relation b/w Bedrooms and Price",fontsize=15)
plt.xlabel("Bedrooms",fontsize=12)
plt.ylabel("Price",fontsize=12)
plt.show()

```

```python
plt.figure(facecolor="#94d2bd")
sns.violinplot(x="floors",y="price",data=df,palette="deep",hue="floors",legend=False)
plt.title("Distribution of price based on Floors",fontsize=15)
plt.xlabel("Floors",fontsize=12)
plt.ylabel("Price",fontsize=12)
plt.gca().set_facecolor("#f8f9fa")
plt.show()
```
